```python
def sampling_func(data, sample_pct):

    np.random.seed(123)

    N = len(data)

    sample_n = int(len(data)*sample_pct) # integer

    sample = data.take(np.random.permutation(N)[:sample_n])

    return sample

```

```python
df2
```

```python
sample_set = df2.groupby('업종명', group_keys=False).apply(sampling_func, sample_pct=0.7)



sample_set.sort_index()


```

```python
test_set = df2.drop(df2.index[sample_set.index])

test_set
```

```python
sample_set['t'] =1
test_set['t'] =0
```

```python
total = pd.concat([sample_set,test_set])
```

```python
total = total.dropna(axis=0)
```

```python
total = total.reset_index(drop=True)
train = total[total.t ==1]
test = total[total.t ==0]
```

```python
train = train.drop('t',axis=1).reset_index()
test = test.drop('t',axis=1).reset_index()
```

```python
train_X = train.drop('이용금액',axis=1)
train_y = train[['index','이용금액']].drop('index',axis=1)
```

```python
test_X = test.drop('이용금액',axis=1)
test_y =test[['index','이용금액']]
```

```python
train_X.to_csv("jeju_train_X.csv",index=0)
train_y.to_csv("jeju_train_y.csv",index=0)
test_X.to_csv("jeju_test_X.csv",index=0)
test_y.to_csv("jeju_test_y.csv",index=0)
```

```python
train_X['target'] =1
test_X['target'] =0
total = pd.concat([train_X,test_X])
```

```python
total['date']  =pd.to_datetime('20'+total['년도'].astype('str') +"-" + total['월'].astype('str')+'-' + total['일'].astype('str'))
```

```python
def weekend(x):
    if x in [5,6]:
        return 1
    else:
        return 0
```

```python
total['weekend']  = total.date.dt.weekday.apply(weekend)
```

```python
dum_col = ['시도명','읍면동명','weekend','업종명']
drop_col = ['index','date','년도','월','일']
```

```python
total[dum_col] = total[dum_col].astype('str')
```

```python
dum_total = pd.get_dummies(total)
```

```python
dum_drop = dum_total.drop(drop_col,axis=1)
dum_train = dum_drop[dum_drop.target==1].drop('target',axis=1)
dum_test = dum_drop[dum_drop.target==0].drop('target',axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(dum_train, train_y.values.reshape(-1,1), test_size=0.3, random_state=42)
```

```python
rf = RandomForestRegressor()
rf.fit(X_train,y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error 
mean_squared_error(y_test, y_pred)
```
